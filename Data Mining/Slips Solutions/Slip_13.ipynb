Q1. Draw a pie chart using R programming for the following datadistribution:[10 Marks]

  Digits on Dice 1 2 3 4 5 6

  Frequency of getting each number 7 2 6 3 4 8



```python
digits <- c(1, 2, 3, 4, 5, 6)
freq <- c(7, 2, 6, 3, 4, 8)
pie(freq, labels = digits, main = "Dice Roll Frequency", col = rainbow(length(freq)))
```
